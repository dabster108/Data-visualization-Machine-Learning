```python
import matplotlib.pyplot as plt
import seaborn as sna
import numpy as np
import pandas as pd
```

1. Starting with matplotlib visualization

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
#plotting sin curve
plt.plot(x,y,label = "Sinusoidal wave")
plt.title("Sinusoidal Wave")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()

```

```python
#Bar chart
categories = ["A", "B", "C" , "D"]
values = [10,20,15,25]
plt.bar(categories,values, color = "violet")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

```

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]
colors = ["violet", "cyan", "orange", "green"]

plt.bar(categories, values, color=colors)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram and Kde plot
import seaborn as sns
data = np.random.randn(1000)
sns.histplot(data, kde = True, color = 'green')
plt.title("Histogram with KDE")
plt.show()

```

```python
#Heat map
matrix = np.random.rand(5,5)
sns.heatmap(matrix,annot =True ,cmap = "coolwarm" ,fmt = ".2f")
plt.title("Heat Map")
plt.show()
```

```python
#paiplot
df = sns.load_dataset("iris")
sns.pairplot(df, hue = "species", markers = ["s","o", "D"])
plt.show()
```
